```python
import spotipy 
from spotipy.oauth2 import SpotifyClientCredentials

client_credentials_manager = SpotifyClientCredentials(client_id=client_id, 
                                                      client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager) #spotify object to access API
```

```python
import pandas as pd
spotify=pd.read_csv('spotify.csv')
df = pd.read_csv("sp.csv")
artist_names = df
print(artist_names.artist)
#print(df)
```

```python
import numpy as np
import json

result_df = pd.DataFrame(columns = ['id','artist','followers','genre','popularity'])
genre_id = []
artist_name = []
follower = []
artist_genre= []
pop= []

for artist in artist_names.artist:
    temp_save = sp.search(q=artist, limit=2, type='artist')
    temp_save = temp_save['artists']['items']
    for x in temp_save:
        #print(x.keys())
        #print(x['name'])
        if x:
            if x['id']:
                genre_id.append(x['id'])
            else:
                genre_id.append("NA")
            if x['name']:
                artist_name.append(x['name'])
            else:
                artist_name.append("NA")
            if x['followers']['total']:
                follower.append(x['followers']['total'])
            else:
                follower.append("NA")
            if x['genres']:
                temp_genre = str(x['genres'])
                temp_genre = temp_genre.replace("[", "")
                temp_genre = temp_genre.replace("]", "")
                temp_genre = temp_genre.replace("'", "")
                artist_genre.append(temp_genre)
            else:
                artist_genre.append("NA")
            if x['popularity']:
                pop.append(x['popularity'])
            else:
                pop.append("NA")

        else:
            genre_id.append("NA")
            artist_name.append("NA")
            follower.append("NA")
            artist_genre.append("NA")
            pop.append("NA")
            
result_df["id"] = genre_id
result_df["artist"] = artist_name
result_df["followers"] = follower
result_df["genre"] = artist_genre
result_df["popularity"] = pop
```

```python
result_df.to_csv("./spotifyoriginal.csv", index = True)
```
